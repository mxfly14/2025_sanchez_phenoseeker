```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Poscon mAP

```python
df_poscon = pd.read_csv('/home/maxime/synrepos/phenospace/normalisation/experiments/dinov2/test_all_transform_2_plates/results/maps_df_poscon.csv')
columns_to_drop = [col for col in df_poscon.columns if 'Random' in col and col != 'Random mAP (raw)']
df_poscon = df_poscon.drop(columns=columns_to_drop)

df_poscon.columns = [col.replace('mAP', '').replace('raw_', '').replace('(', '').replace(')', '').replace('raw', 'Raw').strip() for col in df_poscon.columns]

df_poscon = df_poscon.drop(columns=['Number of Queries'])

df_poscon = df_poscon.set_index('Label')
```

```python
df_example = df_poscon.T.sort_values(by='Mean mAP').T
df_values = df_example.drop(index='Mean mAP').astype(float)
mean_values = df_values.mean()
median_values = df_values.median()
```

```python
df_example = df_example[['Random  Raw', 'Raw', 'ZCA','Z_ZCA', 'Sph', 'Z','ZCA_Z_Int', 'ZCA_Res_Int', 'ZCA_Int',
       'ZCA_Int_Z', 'ZCA_Res_Z_Int', 'Res_ZCA_Z', 'ZCA_Z', 'ZCA_Res_Z']]
df_values = df_example.drop(index='Mean mAP').astype(float)
mean_values = df_values.mean()
median_values = df_values.median()
```

```python
plt.figure(figsize=(12, 6))
for index, row in df_values.iterrows():
    plt.plot(df_values.columns, row, linestyle='', marker='.', label=index)






plt.title('Impact of Normalization Pipelines on each Positive Control Molecule')
plt.ylabel('mAP')
plt.xticks()  # Remove x-axis labels
plt.grid(True)  # Add grid
plt.legend(title='Molecules IDs', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for index, row in df_values.iterrows():
    plt.plot(df_values.columns, row, linestyle='-', label=index)

plt.title('Impact of Normalization Pipelines on each Positive Control Molecule')
plt.ylabel('mAP')
plt.xticks([])  # Remove x-axis labels
plt.grid(True)  # Add grid
plt.legend(title='Molecules IDs', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python

# Step 2: Plot mean values with two dashed lines for specific columns
plt.figure(figsize=(12, 6))
plt.plot(mean_values.index, mean_values, linestyle='', marker='.', color='b', label='Mean mAP Value')

# Add dashed lines for 'raw' and 'Random raw' columns if they exist
if 'Raw' in mean_values.index:
    plt.axhline(y=mean_values['Raw'], color='r', linestyle='--', alpha=0.6, label="Raw Embeddings")
if 'Random  Raw' in mean_values.index:
    plt.axhline(y=mean_values['Random  Raw'], color='g', linestyle='--', alpha=0.6, label="Random Retrieval")

plt.title('Impact of all Normalization Pipelines on Positive Control Molecules Grouping')
plt.ylabel('mAP')
plt.xticks([])  # Remove x-axis labels
plt.grid(True)  # Add grid
plt.legend()
plt.tight_layout()
plt.show()
```
