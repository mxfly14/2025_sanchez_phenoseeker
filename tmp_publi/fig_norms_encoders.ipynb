```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Calculate mean and median for df_poscon_chada
df_example_chada = df_poscon_chada.T.sort_values(by='Mean mAP').T
df_values_chada = df_example_chada.drop(index='Mean mAP').astype(float)
mean_values_chada = df_values_chada.mean()

# Step 2: Calculate mean and median for df_poscon_resnet (duplicate each value to match chada length)
df_example_resnet = df_poscon_resnet.T.sort_values(by='Mean mAP').T
df_values_resnet = df_example_resnet.drop(index='Mean mAP').astype(float)
df_values_resnet = df_values_resnet.loc[df_values_resnet.index.repeat(2)]  # Duplicate each row
mean_values_resnet = df_values_resnet.mean()

# Step 3: Calculate mean and median for df_plates_chada
df_example_plates_chada = df_plates_chada.T.sort_values(by='Mean mAP', ascending=False).T
df_values_plates_chada = df_example_plates_chada.drop(index='Mean mAP').astype(float)
mean_values_plates_chada = df_values_plates_chada.mean()

# Step 4: Calculate mean and median for df_plates_resnet (duplicate each value to match chada length)
df_example_plates_resnet = df_plates_resnet.T.sort_values(by='Mean mAP', ascending=False).T
df_values_plates_resnet = df_example_plates_resnet.drop(index='Mean mAP').astype(float)
df_values_plates_resnet = df_values_plates_resnet.loc[df_values_plates_resnet.index.repeat(2)]  # Duplicate each row
mean_values_plates_resnet = df_values_plates_resnet.mean()

# Step 5: Plot mean values for all datasets
plt.figure(figsize=(24, 12))

# Plot for df_poscon_chada with a different color
plt.plot(range(len(mean_values_chada)), mean_values_chada, linestyle='', marker='o', color='tab:blue', label='ChAdA - Compounds Retrieval')

# Plot for df_poscon_resnet with different color and same marker
resnet_x_values = np.arange(0, len(mean_values_resnet) * 2, 2)  # Spread resnet values across every other x
plt.plot(resnet_x_values, mean_values_resnet, linestyle='', marker='o', color='tab:orange', label='ResNet50 - Compounds Retrieval')

# Plot for df_plates_chada with new symbol
plt.plot(range(len(mean_values_plates_chada)), mean_values_plates_chada, linestyle='', marker='x', color='tab:blue', label='ChAdA - Batch Effect Removal')

# Plot for df_plates_resnet with new symbol and color
plt.plot(resnet_x_values, mean_values_plates_resnet, linestyle='', marker='x', color='tab:orange', label='ResNet50 - Batch Effect Removal')

# Add vertical dashed line for 'Best Normalisation' at the highest mAP value
best_map_value_index = max(mean_values_chada.idxmax(), mean_values_resnet.idxmax(), mean_values_plates_chada.idxmax(), mean_values_plates_resnet.idxmax())
plt.axvline(x=best_map_value_index, color='tab:red', linestyle='--', alpha=0.6, label="No Normalisation")

# Add vertical line at the rightmost x value
plt.axvline(x=len(mean_values_chada) - 1, color='tab:green', linestyle='--', alpha=0.6, label="Best Normalisation")

plt.xticks([])  # Remove x-axis labels
plt.grid(True)  # Add grid

# Move the legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), prop={'size': 20})
plt.tight_layout()
plt.yticks(fontsize=40)
plt.show()
```
