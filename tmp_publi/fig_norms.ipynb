```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_poscon = pd.read_csv('/projects/synsight/repos/phenospace/normalisation/experiments/dinov2/test_all_transform_27_plates/results/maps_df_poscon.csv')
df_plates = pd.read_csv('/projects/synsight/repos/phenospace/normalisation/experiments/dinov2/test_all_transform_27_plates/results/maps_df_controls_Plate.csv')
```

# Poscon mAP

```python
columns_to_drop = [col for col in df_poscon.columns if 'Random' in col and col != 'Random mAP (raw)']
df_poscon = df_poscon.drop(columns=columns_to_drop)

df_poscon.columns = [col.replace('mAP', '').replace('raw_', '').replace('(', '').replace(')', '').replace('raw', 'Raw').strip() for col in df_poscon.columns]

df_poscon = df_poscon.drop(columns=['Number of Queries'])

df_poscon = df_poscon.set_index('Label')
```

```python
df_example = df_poscon.T.sort_values(by='Mean mAP').T
df_values = df_example.drop(index='Mean mAP').astype(float)
mean_values = df_values.mean()
median_values = df_values.median()
```

```python
columns_to_drop = [col for col in df_plates.columns if 'Random' in col and col != 'Random mAP (raw)']
df_plates = df_plates.drop(columns=columns_to_drop)

df_plates.columns = [col.replace('mAP', '').replace('raw_', '').replace('(', '').replace(')', '').replace('raw', 'Raw').strip() for col in df_plates.columns]

df_plates = df_plates.drop(columns=['Number of Queries'])

df_plates = df_plates.set_index('Label')
poscon_index = mean_values.index
df_example_norm = df_plates.T.sort_values(by='Mean mAP', ascending=False).T
df_values_norm = df_example_norm.drop(index='Mean mAP').astype(float)
mean_values_norm = df_values_norm.mean()
median_values_norm = df_values_norm.median()
mean_values_sorted_norm = mean_values_norm.loc[poscon_index]
```

```python

# Step 2: Plot mean values with two dashed lines for specific columns
plt.figure(figsize=(12, 6))
plt.plot(mean_values.index, mean_values, linestyle='', marker='.', color='r', label='Mean mAP Value')

# Add dashed lines for 'raw' and 'Random raw' columns if they exist
#if 'Raw' in mean_values.index:
#    plt.axhline(y=mean_values['Raw'], color='r', linestyle='--', alpha=0.6, label="Raw Embeddings")
#if 'Random  Raw' in mean_values.index:
#    plt.axhline(y=mean_values['Random  Raw'], color='g', linestyle='--', alpha=0.6, label="Random Retrieval")
plt.plot(poscon_index, mean_values_sorted_norm, linestyle='', marker='.', color='b', label='Mean mAP Value')

plt.xticks([])  # Remove x-axis labels
plt.grid(True)  # Add grid
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Plate mAP

```python

# Step 2: Plot mean values with two dashed lines for specific columns
plt.figure(figsize=(12, 6))
plt.plot(poscon_index, mean_values_sorted_norm, linestyle='', marker='.', color='b', label='Mean mAP Value')

# Add dashed lines for 'raw' and 'Random raw' columns if they exist
if 'Raw' in poscon_index:
    plt.axhline(y=mean_values_sorted_norm['Raw'], color='r', linestyle='--', alpha=0.6, label="Raw Embeddings")
if 'Random  Raw' in mean_values_sorted_norm.index:
    plt.axhline(y=mean_values_sorted_norm['Random  Raw'], color='g', linestyle='--', alpha=0.6, label="Random Retrieval")

plt.title('Plate-Level Batch Effect Reduction for Every Normalisation Pipeline')
plt.ylabel('mAP')
plt.xticks([])  # Remove x-axis labels
plt.grid(True)  # Add grid
plt.legend()
plt.tight_layout()
plt.show()
```
